###handling outliers

```python
plt.figure(figsize=(12, 6))
df[['discount_percentage', 'rating', 'rating_count']].boxplot()
plt.title('Box Plot for outliers')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
def find_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

# Call the function
outliers = find_outliers(df['rating_count'])
```

###Labeling

```python
from sklearn.preprocessing import LabelEncoder


product_id = LabelEncoder()
category = LabelEncoder()
review_id = LabelEncoder()
review_content = LabelEncoder()
product_name = LabelEncoder()
user_name = LabelEncoder()
about_product = LabelEncoder()
user_id = LabelEncoder()
review_title = LabelEncoder()
img_link = LabelEncoder()
product_link = LabelEncoder()


df['product_id'] = product_id.fit_transform(df['product_id'])
df['category'] = category.fit_transform(df['category'])
df['review_id'] = review_id.fit_transform(df['review_id'])
df['review_content'] = review_content.fit_transform(df['review_content'])
df['product_name'] = product_name.fit_transform(df['product_name'])
df['user_name'] = user_name.fit_transform(df['user_name'])
df['about_product'] = about_product.fit_transform(df['about_product'])
df['user_id'] = user_id.fit_transform(df['user_id'])
df['review_title'] = review_title.fit_transform(df['review_title'])
df['img_link'] = img_link.fit_transform(df['img_link'])
df['product_link'] = product_link.fit_transform(df['product_link'])
```

###Skewness


```python
import seaborn as sns
for column in df.columns:
    skewness = df[column].skew()
    kurtosis = df[column].kurtosis()
    print(f"{column}:")
    print(f"  Skewness: {skewness:.2f}")
    print(f"  Kurtosis: {kurtosis:.2f}")

    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True, color='blue', bins=20, alpha=0.7)
    plt.title(f"Distribution of {column}\nSkewness: {skewness:.2f}, Kurtosis: {kurtosis:.2f}", fontsize=16)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
```

```python
from textblob import TextBlob

df_reviews = df[["product_id", "user_id", "review_content"]]


df_reviews["sentiment"] = df_reviews["review_content"].map(lambda text: TextBlob(text).sentiment.polarity)


most_positive_reviews = df_reviews.sort_values(by="sentiment", ascending=False).head(10)


most_negative_reviews = df_reviews.sort_values(by="sentiment", ascending=True).head(10)


print("Top 10 Positive Reviews:")
print(most_positive_reviews)

print("\nTop 10 Negative Reviews:")
print(most_negative_reviews)
```
